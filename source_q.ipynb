```python
import re
import random
def Pretreat(file):
    docs = []
    file = open(file, 'r',  encoding='UTF8')
    docs = file.read().replace('\n\n','\n')
    doc= docs.split('\n')
    
    return doc

def test1(file, num):
    ques=''
    an=[]
    ex=[]
    doc = Pretreat(file)
    
    
    i=0
    for line in doc:
        temp = re.search("( for | if | else )",line)
        if temp :
            blank='('+ ' '* int(len(line)/2)+ chr(i+65) + ' '* int(len(line)/2)  +')'
            ques+=' '*line.find(temp[0])+temp[0]+blank+'\n'
            
            an.append( line[ (line.find(temp[0])+ len(temp[0])) :] )
            i+=1
        else:
            ques+=line+"\n"
             
    print("빈칸에 들어갈 답을 고르시오.\n")
    print('코드')
    print(ques)
    print('보기')
    ex = an.copy()
    random.shuffle(ex)
    for i in range(len(ex)):
        alpa = chr(i+65)
        print( alpa + ' : ' + ex[i])
    return ques, an , ex
    
    
def test2(file):
    ques=''
    an=[]
    ex=[]
    an_a=[]
    
    doc = Pretreat(file)
    i=0
    while  i  < len(doc) :
        line = doc[i]
        temp = re.search("( for | while)",line)
        if temp :
            b = line.rfind('    ')    
            blank=' '*(b+4)+'('+' '*len(line)+')'    
            while b <= line.rfind('    ')  :
                ques+= blank+'\n'
                an.append(line.lstrip())
                i+=1                       
                line = doc[i]
        else:
            ques+=line+"\n"
            i+=1
            
    print("빈칸에 들어갈 소스코드의 순서를 맞게 재배열하시오.\n")
    print('코드')
    print(ques)
    print('보기')
    
    ex = an.copy()
    random.shuffle(ex)
    for i in range(len(an)):
        alpa = chr(i+65)
        print( alpa + ':' + ex[i])
        an_a.append( chr(ex.index( an[i])+ 65 ) )
        
    return ques, an , an_a
```
